```python
import geopandas as gpd
import pandas as pd
import folium
```

```python
rj = gpd.read_file('/content/drive/MyDrive/Colab Notebooks/Python 16 - Geopandas pt1/RJ/RJ-MUNIC/RJ-MUNIC.shp')
geo_dados = gpd.read_file('drive/MyDrive/Colab Notebooks/Python 16 - Geopandas pt1/RJ/DATASET/DATASET.shp')
setor= gpd.read_file('drive/MyDrive/Colab Notebooks/Python 16 - Geopandas pt1/RJ/DATASET/DATASET.shp')
bairro = gpd.read_file('/content/drive/MyDrive/Colab Notebooks/Python 16 - Geopandas pt1/RJ/RJ-BAIRRO/RJ-BAIRRO.shp')
```

```python
crs = {'init': 'epsg:4326'}
```

```python
setor = setor.fillna(0)
```

```python
rj.to_crs(crs, inplace = True)
```

```python
rj.centroid
```

```python
y = rj.centroid.y.iloc[0]
```

```python
x = rj.centroid.x.iloc[0]
```

```python
rj
```

#Mapa Camada Simples

```python
base = folium.Map([y,x], zoom_start=11, tiles = 'OpenStreetMap')
base.choropleth(rj)
base
```

```python
base = folium.Map([y,x], zoom_start=11, tiles = 'OpenStreetMap')
geojson_rj = folium.GeoJson(rj)
geojson_rj.add_child(folium.Popup(rj.NM_MUNICIP.iloc[0])) #Adiciona PopUp
geojson_rj.add_to(base)
base
```

#Mapa Várias Camadas

```python
bairro.head(1)
```

```python
bairro.to_crs(crs, inplace= True)
```

```python
base = folium.Map([y,x], zoom_start=11, tiles = 'OpenStreetMap')
for i in range(len(bairro)):
  geo = folium.GeoJson(bairro[i:i+1], name = bairro['NM_BAIRRO'][i])
  label = '{} - {} habitante'.format(bairro['NM_BAIRRO'][i], bairro['V002'][i])
  folium.Popup(label).add_to(geo)
  geo.add_to(base)

folium.LayerControl().add_to(base)
```

```python
base
```

```python
base.save('Bairros (RJ).html')
```

##Exercício

```python
base = folium.Map([y,x], zoom_start=11, tiles = 'OpenStreetMap')
for i in range(len(bairro)):
    geo = folium.GeoJson(
        bairro[i:i+1],
        name=bairro['NM_BAIRRO'][i],
        style_function=lambda feature: {
    'fillColor': 'green' if feature['properties']['V002'] >= bairro['V002'].mean() else 'yellow',
    'fillOpacity': 0.3,
    'color': 'green' if feature['properties']['V002'] >= bairro['V002'].mean() else 'yellow',
    'weight': 1,
    'opacity': 0.1
}
    )
    label = '{} - {} habitantes'.format(bairro['NM_BAIRRO'][i], bairro['V002'][i])
    folium.Popup(label).add_to(geo)
    geo.add_to(base)
base
```

##Mapa com Marcadores

```python
base = folium.Map([y, x], zoom_start=11, tiles='OpenStreetMap')
base.choropleth(rj, name = 'Rio de Janeiro', line_color='Black', line_weight=3, fill_opacity=0)

amostra = geo_dados.sample(500)
 
for item in amostra.itertuples():
  base.add_child(folium.Marker(location=[item.Latitude, item.Longitude], 
                               popup = '<h4>'+str(item.Tipo)+'<h4><h5>'+str(item.Bairro)+'<h5></p>R$'+'{:,}'.format(item.Valor)+'</p>',
                               icon = folium.Icon(color = 'orange' if item.Valor>1000000 else 'red', prefix='fa', icon='fas fa-home'))) #PAGE de ICONES https://fontawesome.com/v5.15/icons
folium.LayerControl().add_to(base)

base.save('Dataset Marcadores.html')
base
```

##Mapa com Marcadores Clusterizados

```python
from folium.plugins import MarkerCluster
```

```python
brt = gpd.read_file('drive/MyDrive/Colab Notebooks/Python 16 - Geopandas pt1/Transporte/BRT/Esta%C3%A7%C3%B5es_BRT.geojson')
```

```python
brt.head(1)
```

```python
brtgeo = folium.GeoJson(brt)
```

```python
base = folium.Map([y, x], zoom_start=11, tiles='OpenStreetMap', control_scale=True,)
base.choropleth(rj, name = 'Rio de Janeiro', line_color='Black', line_weight=3, fill_opacity=0)

cluster = MarkerCluster()
 
for item in geo_dados.itertuples():
  cluster.add_child(folium.Marker(location=[item.Latitude, item.Longitude], 
                               popup = '<h4>'+str(item.Tipo)+'<h4><h5>'+str(item.Bairro)+'<h5></p>R$'+'{:,}'.format(item.Valor)+'</p>',
                               icon = folium.Icon(color = 'orange' if item.Valor>1000000 else 'red', prefix='fa', icon='fas fa-home'))) #PAGE de ICONES https://fontawesome.com/v5.15/icons
base.add_child(cluster)

folium.LayerControl().add_to(base)
folium.plugins.measure_control.MeasureControl().add_to(base)
folium.plugins.MiniMap().add_to(base)
folium.plugins.ScrollZoomToggler().add_to(base)
folium.plugins.Search(layer=brtgeo, search_label='Nome', collapsed=False).add_to(base) #Não Funciona

base.save('Dataset Marcadores Cluster.html')
base
```
