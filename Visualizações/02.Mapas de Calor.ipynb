```python
import geopandas as gpd
import folium
```

```python
rj = gpd.read_file('/content/drive/MyDrive/Colab Notebooks/Python 16 - Geopandas pt1/RJ/RJ-MUNIC/RJ-MUNIC.shp')
geo_dados = gpd.read_file('drive/MyDrive/Colab Notebooks/Python 16 - Geopandas pt1/RJ/DATASET/DATASET.shp')
setor= gpd.read_file('drive/MyDrive/Colab Notebooks/Python 16 - Geopandas pt1/RJ/RJ-SETOR/RJ-SETOR.shp')
bairro = gpd.read_file('/content/drive/MyDrive/Colab Notebooks/Python 16 - Geopandas pt1/RJ/RJ-BAIRRO/RJ-BAIRRO.shp')
```

```python
crs = {'init': 'epsg:4326'}
```

```python
rj = rj.to_crs(crs)
geo_dados = geo_dados.to_crs(crs)
setor = setor.to_crs(crs)
bairro = bairro.to_crs(crs)
```

```python
setor.fillna(0, inplace = True)
```

```python
y = rj.centroid.y.iloc[0]
x = rj.centroid.x.iloc[0]
```

```python
setor.head(1)
```

```python
data = []
for i in range (len(setor)):
  data.append([setor['geometry'][i].centroid.y, setor['geometry'][i].centroid.x, setor['V002'][i]/setor['V002'].max()])
```

```python
data
```

```python
from folium.plugins import HeatMap
base = folium.Map([y,x], tiles="OpenStreetMap", zoom_start=11)

HeatMap(data, name = 'Popula√ß√£o').add_to(base)

folium.LayerControl().add_to(base)

base.save('Dataset Marcadores HeatMap.html')
base
```
